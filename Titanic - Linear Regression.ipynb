```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')
titanic.head(10)
```

```python
titanic.describe()
```

```python
titanic.drop(['PassengerId', 'Ticket', 'Name'], axis = 1, inplace=True)
```

```python
titanic.head()
```

```python
titanic.groupby('Survived').mean()
```

```python
titanic.groupby('Embarked').mean()
```

```python
titanic.isnull().sum()
```

```python
titanic.groupby(titanic['Age'].isnull()).mean()
```

```python
average = titanic['Age'].mean()
titanic['Age'].fillna(value = average, inplace = True)
```

```python
titanic.isnull().sum()
```

```python
titanic.head()
```

```python
titanic['Family_count'] = titanic['SibSp'] + titanic['Parch']
```

```python
titanic.head()
```

```python
titanic.drop(labels = ['SibSp', 'Parch'], axis = 1, inplace=True)
```

```python
titanic.head()
```

```python
import numpy as np
titanic['Cabin_indicator'] = np.where(titanic['Cabin'].isnull(), 0, 1)
titanic.head()
```

```python
titanic.drop(['Cabin', 'Embarked'], axis = 1, inplace = True)
```

```python
titanic.head()
```

```python
gender_num = {'male': 0, 'female':1}

titanic['Sex'] = titanic['Sex'].map(gender_num)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
labels = titanic['Survived']
features = titanic.drop('Survived', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier()

parameters = {
    'n_estimators': [5, 20, 40, 80, 100],
    'max_depth': [2, 5, 10, None],
    'min_samples_split': [2, 3, 5],
    'criterion':['gini', 'entroy', 'log_loss']
}

cv = GridSearchCV(rf, parameters, cv = 5)
```

```python
print(5*4*3*3*5)
```

```python
cv.fit(X_train, y_train.values.ravel())
```

```python
best_model = cv.best_estimator_
```

```python
print(cv.best_params_)
```

```python
best_model.predict(X_test)
```

```python
means = cv.cv_results_['mean_test_score']
stds = cv.cv_results_['std_test_score']

for mean, std, params in zip(means, stds, cv.cv_results_['params']):
  print('{} (+/-{}) for {}'.format(round(mean,3), round(std*2, 3), params))
```
